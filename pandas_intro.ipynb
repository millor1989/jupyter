pandas 的两个核心对象：DataFrame 和 Series

#### 1、 DataFrame

DataFrame 是一个表，它包含了一个单元格阵列（an array of individual entries），每个单元都有一个值。每个单元对应于一行和一列。

```python
pd.DataFrame({'Yes':[50, 21], 'No':[131, 2]})
```

“0,No”对应的单元格为 131，“0，Yes”对应的单元格为 50。

DataFrame 单元不一定是整数，还可以是字符串等类型：

```python
pd.DataFrame({'Bob': ['I liked it.', 'It was awful.'], 'Sue': ['Pretty good.', 'Bland.']})
```

之前两个例子中使用了 `pd.DataFrame()` 构造器来生成 DataFrame 对象。

语法是：声明一个 dict 键为列名，值为单元格值的集合。

这个字典-集合构造器为列标签赋了值，但是对于行标签使用的是从 0 开始的升序计数。如果要指定行标签，可以使用构造器的 `index` 参数。



```python
pd.DataFrame({'Bob': ['I liked it.', 'It was awful.'], 
              'Sue': ['Pretty good.', 'Bland.']},
             index=['Product A', 'Product B'])
```

#### 2、Series

Series 是一个数值的序列。DataFrame 可以看作一个表，而 Series 可以看作一个集合。

使用集合创建 Series：

```python
pd.Series([1, 2, 3, 4, 5])
```

本质上，Series 是 DataFrame 的一列。

与之前类似，可以使用 `index` 参数为 Series 指定行标签。但是 Series 没有列名（列是对 DataFrame 而言的），只有名字 `name`。

```python
pd.Series([30, 35, 40], index=['2015 Sales', '2016 Sales', '2017 Sales'], name='Product A')
```

**Indexing in pandas**

pandas自有的访问操作符`loc`和`iloc`

1、index-based selection `iloc`基于（numerical position）位置进行数据选择

2、label-based selection `loc`基于（index value）索引（列名）进行数据选择

`loc`和`iloc`的第一个参数是面向行的，第二个参数是面向列的，这正好和Python相反。

1）可以使用`:`操作符作为参数——表示“所有”。`:`和数值一起使用时表示一个区间（**注意**`iloc`是前闭后开区间，`loc`是前闭后闭的区间）；

2）可以使用集合作为参数；

3）可以使用负数——从后往前进行选择

```python
# 获取第一行数据
reviews.iloc[0]
```

```python
# 获取第一列数据
reviews.iloc[:, 0]
```

```python
# 0~2行的第一列
reviews.iloc[:3, 0]
```

```python
# 第2行和第3行的第一列
reviews.iloc[1:3, 0]
```

```python
# 使用集合作参数
reviews.iloc[[1,2,3],0]
```

```python
# 选择末尾5行
reviews.iloc[-5:,0]
```

```python
reviews.loc[0, 'country']
```

```python
reviews.loc[:, ['taster_name', 'taster_twitter_handle', 'points']]
```

`loc` 是闭区间：

这样设计是为了应对索引非数值的场合

```python
reviews.loc[:1, ['taster_name', 'taster_twitter_handle', 'points']]
```

**Maps**(映射)

映射是一个从数学上借用的概念，用一个函数把一个值的集合映射为另一个值的集合。

1）`map()`函数，从Series获取一个值并返回经过指定函数转换后的值，返回结果是一个新的Series

2）`apply()`函数，对DataFrame的每行（`axis='columns'`）或每列（`axis='index'`）进行转换,与`map()`类似，根据函数返回值返回一个新的DataFrame或Series；会比`map()`慢

3）Pandas还有很多内置的映射操作，比如，数值Series与固定数值或者数值Series（两个Series的长度相同）的数学运算；比如，用字符串连接Series；Pandas Series几乎支持所有的Python运算符（比如`>`、`<`、`==`等等）。这些内置映射操作比`map()`、`apply()`要快；

```python
review_points_mean = reviews.points.mean()
reviews.points.map(lambda p: p - review_points_mean)
```

```python
def remean_points(row):
    row.points = row.points - review_points_mean
    return row

reviews.apply(remean_points, axis='columns')
```

`map()` 和 `apply()` 都是返回新的、转换后的 Series 和 DataFrame。不会修改调用它们的原始数据。

```python
reviews.head(1)
```

Pandas 提供了很多内置的 mapping 操作。

如下例子中，表面上是 Series 和一个数值的减法，实际上是为 Series 的每一个值分别执行与给定数值的减法运算

```python
review_points_mean = reviews.points.mean()
reviews.points - review_points_mean
```

```python
reviews.country + " - " + reviews.region_1
```

这些操作比 `map()` 或 `apply()` 还要快，是因为 Pandas 内置了提速优化。

虽然快，但是不如 `map()` 或 `apply()` 灵活，比如，在需要进行复杂逻辑运算时。

```python
def stars(row):
    if row.country == 'Canada':
        return 3
    elif row.points >= 95:
        return 3
    elif row.points >= 85:
        return 2
    else:
        return 1

reviews.apply(stars, axis='columns')
```

```python
reviews.country + " - " + reviews.region_1
```

**分组和排序**

1、`groupby()`,分组

1）可以对超过一个字段进行分组

2）分组后可以使用任意的汇总函数（summary functions），并且可以使用`agg()`函数运行一系列不同的汇总函数

3）分组后也可以使用`apply()`函数进行处理，每组可以被当作DataFrame一个子DataFrame，传入`apply()`的处理函数以分组为单位对数据进行操作

2、多索引（Multi-indexes）

多索引与单个索引不同于：它有多个层级（multiple levels）。对于多索引的数据集需要用多索引才能获取到一个值。但是，多索引最常用的方法是`reset_index()`：将多索引转换为单个索引。

3、排序（Sorting）

`groupby`按照索引的值的顺序而不是数据的值的顺序返回结果。

1）`sort_values()`函数按值进行排序（默认升序）。可以按照多个值排序

2）`sort_index()`函数按索引进行排序，用法与`sort_values()`类似

```python
# 可以使用groupby实现`value_counts()`的功能
reviews.groupby('points').points.count()
```

```python
# 每组points中price最低的记录
reviews.groupby('points').price.min()
```

```python
# 每组winery中第一次出现的titile
reviews.groupby('winery').apply(lambda df: df.title.iloc[0])
```

```python
# 按照country和province分组，查找每组points最大的记录
reviews.groupby(['country', 'province']).apply(lambda df : df.loc[df.points.idxmax()])
```

```python
# 同时进行len、min、max统计
reviews.groupby('country').price.agg([len, min, max])
```

```python
# 多字段groupby返回的就是一个多索引（Multi-indexes）DataFrame
countries_reviewed = reviews.groupby(['country', 'province']).description.agg([len])
countries_reviewed
```

```python
# 多索引的类型
mi = countries_reviewed.index
type(mi)
```

```python
countries_reviewed.reset_index()
```

```python
countries_reviewed = countries_reviewed.reset_index()
countries_reviewed.sort_values(by='len')
```

```python
# 降序排序
countries_reviewed.sort_values(by='len', ascending=False)
```

```python
# 多值排序
countries_reviewed.sort_values(by=['country', 'len'])
```

```python
# 索引排序
countries_reviewed.sort_index()
```

**Combining**

1）`concat`，沿指定的轴整合

2）`join`，连接两个拥有相同索引的DataFrame，可以为左表或者右表的字段设置后缀（如果进行连接的两个表具有相同的列名，那么列后缀是必须指定的）

```python
canadian_youtube = pd.read_csv("./datasets/youtube-new/CAvideos.csv")
british_youtube = pd.read_csv("./datasets/youtube-new/GBvideos.csv")

pd.concat([canadian_youtube, british_youtube])
```

```python
left = canadian_youtube.set_index(['title', 'trending_date'])
right = british_youtube.set_index(['title', 'trending_date'])

left.join(right, lsuffix='_CAN', rsuffix='_UK')
```
